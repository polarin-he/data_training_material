# (Meta)data should be assigned a globally unique and persistent identifier (F1)

This notebook gives a detailed description of the FAIR guiding princple F1: *(Meta)data should be assigned a globally unique and persistent identifier.* Principle F1 is arguably the most important because it will be hard to achieve other aspects of FAIR without globally unique and persistent identifiers. Hence, compliance with F1 will already take you a long way towards publishing FAIR data.

The easiest way to explain the F1 FAIR guiding principles is by explaining terms mentioned in the statement individually: *globally unique*, *persisent* and *identifiers*. Let's first elaborate on that last term: *identifiers*. 

## What is an identifier?
An identifier is a label used to uniquely distinguish an object, entity, or piece of information within a specific context. Identifiers can help other people understand exactly what you mean, and they allow computers to interpret your data in a meaningful way (i.e., computers that are searching for your data or trying to automatically integrate them). Identifiers are essential to the human-machine interoperation that is key to the vision of Open Science (which is explained in more detail in [this notebook](../00_Introduction/03_FAIR_in_perspective.ipynb)). In addition, identifiers will help others to properly cite your work when reusing your data.

## What makes an identifier *globally unique*?
Globally unique identifiers (GUIDs) - often refered to as Universally Unique Identifiers (UUIDs) - are identifiers that are designed to be unique across all namespaces. This means that the identifier should be unique across all systems and databases. This allows for the identifier to be used in different systems and databases without the risk of ambiguity. 

### But how unique is unique? 
Even though the likelihood of two items of information across the internet share the same GUID, the odds of that happening are not zero. Assuming a randomly generated GUID you have about the same chance of getting hit by a meteorite in a year as getting a collision in 10-30 trillion GUIDs. 

When generated according to the standard methods, UUIDs are, for practical purposes, unique. Their uniqueness does not depend on a central registration authority or coordination between the parties generating them, unlike most other numbering schemes. While the probability that a UUID will be duplicated is not zero, it is generally considered close enough to zero to be negligible.

### How do GUIDs work?
GUIDs are constructed in a sequence of digits that equal 128 bits. The ID is in hexadecimal digits, meaning it uses the numbers 0 through 9 and letters A through F. The hexadecimal digits are grouped in a format that is 36 characters long -- 32 hexadecimal characters grouped as 8-4-4-4-12 and separated by four hyphens: {XXXXXXXX-XXXX-XXXX-XXXX-XXXXXXXXXXXX}.

### What's so great about GUIDs?
- No need for a central authority to ensure uniqueness
- You probably wont run out - there's abou  t 75,000,000,000,000,000,000 grains of sand on earth but even that pales in comparision to the number of GUIDs available: 340,282,366,920,938,463,463,374,607,431,770,000,000
- Easy to merge: You can merge different datasets as GUID entity identifiers are very unlikely to collide

### How do I make a UUID or GUID for my data?
To create a UUID or GUID, you can use programming libraries that provide built-in support for generating them. For example, in Python, the `uuid` module can be used to generate UUIDs. By calling `uuid.uuid4()`, you can create a random UUID based on random numbers, ensuring a high degree of uniqueness. This method is simple and does not require any external dependencies or central authority.

```python
# Try it for yourself!
import uuid
generated_uuid = uuid.uuid4()
print(f"Generated UUID: {generated_uuid}")
```

## What makes an identifier *persistent*?
Because it takes time and money to keep web links active, web links tend to become invalid over time. To ensure persistence, identifiers are often managed by trusted organizations or registry services that commit to maintaining the resolvability of the identifier over time. This means that even if the underlying data or metadata changes location or ownership, the identifier will still reliably point to the correct resource. Persistent identifiers (PIDs) are crucial for long-term data accessibility and reproducibility, as they prevent issues like "link rot" where resources become unreachable due to broken links. 

## So why are *globally unique* and *persistent* identifiers important?
Globally unique and persistent identifiers remove ambiguity in the meaning of your published data by assigning a unique identifier to every element of metadata and every concept/measurement in your dataset. 

Many data repositories will automatically generate globally unique and persistent identifiers to deposited datasets, and tag these datasets accordingly in the accompanied metadata. Depending on the type of data however, it could be helpful to assign identifiers to individual data elements (observation or measurement) or concepts within the dataset. This will allow others to refer to specific parts of your data, which can be especially useful when your dataset is large or complex.

## Real-world examples of *globally unique* and *persistent* identifiers 
Here are some examples of identifiers that are both **persistent** and **globally unique**:

- **DOI (Digital Object Identifier):**  
  Used to uniquely identify digital objects such as research papers, datasets, and reports. Here is an identifier that uniquely links to a synthesis paper about the SESS recommendations from the Svalbard Integrated Arctic Earth Observing System (SIOS): `10.5281/zenodo.13322894` (directly accesible via https://doi.org/10.5281/zenodo.13322894)

- **ORCID (Open Researcher and Contributor ID):**  
  A persistent identifier for researchers, ensuring their work is correctly attributed regardless of name changes or variations. One particular person on planet earth has this globally unique and persistent identifier: `0009-0005-8856-8497` (directly accesible via https://orcid.org/0009-0005-8856-8497)

- **UUID (Universally Unique Identifier):**  
  Commonly used in software systems to uniquely identify objects or entities. A specific human observation of the Rangifer tarandus species (more popularly known as the *Svalbard Reindeer*) has a UUID given by the Biodiversity Information Facility (GBIF): `1a97839a-a457-434c-9706-a48f79bfe2fb` (directly accesible via https://www.gbif.org/occurrence/5034875638)

- **ISBN (International Standard Book Number):**  
  A unique identifier for books, ensuring that specific editions can be reliably referenced. Here is an ISBN that links to one of the SESS chapter of the 2024 SESS report by SIOS: `978-82-93871-14-9`.

- **ARK (Archival Resource Key):**  
  Used for persistent access to digital objects in archives. An example is the ARK identifier for the famous _La Joconde_ painting by Léonard de Vinci: `ark:/53355/cl010066723` (directly accesible via https://n2t.net/ark:/53355/cl010066723)

- **Handle System Identifiers:**  
  Similar to DOIs, used for persistent identification of digital resources. An example is this description of the CNRI License Agreement for Proxy Servlet: `20.1000/100` (directly accesible via https://hdl.handle.net/20.1000/100)

- **Other proprietary identifiers:**
  The human polycystin-1 protein has a globally unique and persistent identifier given by the UniProt database, namely `P98161` (directly accesible via http://purl.uniprot.org/uniprot/P98161)

These identifiers are designed to remain resolvable and reliable over time while ensuring global uniqueness across systems and contexts.
