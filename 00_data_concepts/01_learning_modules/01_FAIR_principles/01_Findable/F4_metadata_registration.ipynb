# (Meta)data should be registered or indexed in a searchable resource (F4)

This notebook gives a detailed description of the FAIR guiding princple F4: *(Meta)data should be registered or indexed in a searchable resource.* Even though the (meta-)data adheres to the FAIR principles F1 - F3, the data might not be discoverable (by human or machine) if it is not registered or indexed in a searchable resource. The data might therefore never be used, which would be a waste of resources.
