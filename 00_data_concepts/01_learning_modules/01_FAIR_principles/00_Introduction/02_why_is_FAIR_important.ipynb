# So why should I _care_ about _FAIR_?!
Many researchers are not aware of the importance of FAIR data. They see it as an extra burden, as something that is not relevant to their research, or as something that is only important for data-intensive research. 

The reality is, **most scientific data never reach their full potential to drive new discoveries**. [Frontiers (2025)](https://www.sciencedaily.com/releases/2025/10/251013040314.htm)<sup>1</sup> found that out of every 100 datasets produced, about 80 stay within the lab, 20 are shared but seldom reused, fewer than two meet FAIR standards, and only one typically leads to new findings. Other studies<sup>2,3</sup> suggest that **up to 80% of scientific research data are lost within 20 years**. And that estimate does not even take into account the effects of practical unavailability of such data, when they have not been robustly and accessibly archived, even when they are not actually “lost”. 
 
**The consequences are significant**: slower progress in cancer treatment, climate models that lack sufficient evidence, and studies that cannot be replicated. The steady loss of usable data blocks the ability to validate results and reuse information in science.

This disconnect — between how researchers perceive FAIR data and the stark reality of its impact — is not just a missed opportunity; it is a systemic barrier to scientific progress. When data are not findable, accessible, interoperable, or reusable, their value erodes over time, leaving gaps in knowledge that could otherwise accelerate breakthroughs. FAIR principles address these challenges head-on by making data actionable, not just for today’s researchers, but for future innovators. When data are FAIR, they become a shared resource, enabling:

- **Reproducibility**: Studies can be verified and extended, reducing wasted effort on dead-end paths.
- **Interdisciplinary collaboration**: Climate scientists, medical researchers, and engineers can integrate datasets from different sources and disciplines so that they can feed into decision support systems and dashboards and be used by authorities and politicians to tackle complex societal challenges like pandemics or energy crises. By publishing data in a FAIR-compliant manner, you can contribute to the development of such services for the benefit of society! Some examples are given in [this notebook](../../../02_additional_resources/01_supporting_literature.ipynb).
- **Long-term impact**: Data outlive individual projects, fueling discoveries decades later—just as historical datasets now underpin AI advancements.

The consequences of poor data management are clear: lost opportunities, unreproducible research, and stalled progress. But as [Wilkinson et al (2016)](https://doi.org/10.1038/sdata.2016.18) emphasize, **Good data management is not a goal in itself, but rather is the key conduit leading to knowledge discovery and innovation, and to subsequent data and knowledge integration and reuse by the community.** This is precisely what the FAIR principles to achieve. FAIR transforms data from isolated outputs into active assets for the scientific community. By ensuring data are well-described, accessible, and compatible across disciplines, FAIR principles turn Wilkinson’s vision into practice:

- **Findable and Accessible**: Data are no longer trapped in silos, enabling researchers to locate and build on existing work.
- **Interoperable and Reusable**: Datasets become interlinked, allowing integration across studies and fields just as Wilkinson’s "conduit" metaphor suggests.

>NOTE:
The FAIR Data Principles are central to the [Horizon Europe 2020 guidelines on FAIR data management](https://ec.europa.eu/research/participants/data/ref/h2020/grants_manual/hi/oa_pilot/h2020-hi-oa-data-mgt_en.pdf) that POLARIN adheres to. This document also provides tips on how to create a Data Management Plan (DMP), which we will cover in [this notebook](../../../../01_data_collection/01_learning_modules/01_project_preparation/01_data_management_plans.ipynb). [The European Code of Conduct for Research Integrity (2017)](https://icac.cat/wp-content/uploads/2017/04/codi_allea_17.pdf) also expects that access to research data is in line with the FAIR principles where appropriate.

### Summary
In short, FAIR should not be seen as an administrative hurdle; it’s about unlocking the full potential of science itself. It is the foundation for ensuring that data can be discovered, validated, and built upon by others. Without it, the cycle of lost potential continues: datasets gather digital dust, collaborations stall, and critical questions go unanswered.

So, if you want your research to be relevant, either now or in 20 years from now, publish your research data in a FAIR-compliant manner! Bonus points if you do so directly after collecting them.
