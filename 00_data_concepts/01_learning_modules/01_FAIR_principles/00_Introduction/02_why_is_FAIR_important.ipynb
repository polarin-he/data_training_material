# Why is FAIR data important?
Many researchers are not aware of the importance of FAIR data. They see it as an extra burden, as something that is not relevant to their research, or as something that is only important for data-intensive research. However, **FAIR data is important for all research disciplines and all types of research data**.

**Your research data are valueable by their own right**. Just like your publications, the research data itself is a result of your research efforts and can be used to answer new research questions, to validate and reproduce research results, to develop new methods and tools, to create new knowledge, and to inform policy and practice.

**Publishing data should therefore be done seperately from publishing papers**. In the interest of *open science*, data can be published well before the publication of any paper, ideally as soon as it has been collected, processed, and quality controlled.  

**FAIR data can benefit society by integration into much larger services and platforms**. By publishing data in a FAIR-compliant manner, you can contribute to the development of new services and platforms that integrate data from different sources and disciplines. This can lead to new insights and discoveries, and can help to address societal challenges. Some examples are given in [this notebook](../../../02_additional_resources/01_supporting_literature.ipynb).

**Recent publications suggest that up to 80% of scientific research data are lost within 20 years**. And that estimate does not even take into account the effects of practical unavailability of such data, when they have not been robustly and accessibly archived, even when they are not actually “lost”. Data attrition blocks the ability to validate results and reuse information in science.

[Wilkinson et al (2016)](https://doi.org/10.1038/sdata.2016.18) summarizes the value of good data management effectively: **"Good data management is not a goal in itself, but rather is the key conduit leading to knowledge discovery and innovation, and to subsequent data and knowledge integration and reuse by the community after the data publication process."**
